### I’m subsetting and aggregating dataset at daily basis to explain the different methods.

* Subsetting the dataset from (August 2012 – Dec 2013)
* Creating train and test file for modeling. The first 14 months (August 2012 – October 2013) are used as training data and next 2 months (Nov 2013 – Dec 2013) as testing data.
* Aggregating the dataset at daily basis 

```python
# Subsetting the dataset
# Index 11856 marks the end of year 2013
df = pd.read_csv('data/train.csv', nrows = 11856)
```

```python
# Creating train and test set 
# Index 10392 marks the end of October 2013 
train = df[0 : 10392] 
test = df[10392 : 11856]
```

```python
#Aggregating the dataset at daily level
df.Timestamp = pd.to_datetime(df.Datetime, format='%d-%m-%Y %H:%M')
df.index = df.Timestamp
df = df.resample('D').mean()

train.Timestamp = pd.to_datetime(train.Datetime, format='%d-%m-%Y %H:%M') 
train.index = train.Timestamp 
train = train.resample('D').mean() 

test.Timestamp = pd.to_datetime(test.Datetime, format='%d-%m-%Y %H:%M') 
test.index = test.Timestamp 
test = test.resample('D').mean()
```

```python
dd = np.asarray(train.Count)
```

## Illustration of Different Time Series Model

### Naive Method

```python
y_hat = test.copy()
y_hat['naive'] = dd[len(dd)-1]
```

```python
train['Count'].plot(figsize = (12, 5), title = 'Daily Ridership', fontsize = 15)
test['Count'].plot(figsize = (12, 5))
y_hat['naive'].plot(figsize = (12, 5))
```

```python
plt.figure(figsize = (12,8))
plt.plot(train.index, train['Count'], label = 'Train')
plt.plot(test.index, test['Count'], label = 'Test')
plt.plot(y_hat.index, y_hat['naive'], label = 'Naive Forecast')
plt.legend(loc = 'best')
plt.title("Naive Forecast")
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Count, y_hat.naive))
print(rms)
```

### Simple Average

```python
y_hat_avg = test.copy()
y_hat_avg['avg'] = train['Count'].mean()
```

```python
plt.figure(figsize=(12,8))
plt.plot(train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_avg['avg'], label = 'Average Forecast')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Count, y_hat_avg.avg))
print(rms)
```

### Moving Average

```python
y_hat_mavg = test.copy()
y_hat_mavg['mavg'] = train['Count'].rolling(60).mean().iloc[-1]
```

```python
plt.figure(figsize=(12,8))
plt.plot(train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_mavg['mavg'], label = 'Moving Average Forecast')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Count, y_hat_mavg.mavg))
print(rms)
```

### Simple Exponential Smoothing

```python
y_hat_SEavg = test.copy()
fit2 = SimpleExpSmoothing(np.asarray(train['Count'])).fit(smoothing_level=0.6, optimized=False)
y_hat_SEavg['SES'] = fit2.forecast(len(test))
```

```python
plt.figure(figsize=(16,8))
plt.plot(train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_SEavg['SES'], label='SES')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Count, y_hat_SEavg.SES))
print(rms)
```

### Holt's Linear Trend Model

```python
sm.tsa.seasonal_decompose(train.Count).plot()
result = sm.tsa.stattools.adfuller(train.Count)
plt.show()
```

```python
y_hat_HTavg = test.copy()

fit1 = Holt(np.asarray(train['Count'])).fit(smoothing_level = 0.3, smoothing_slope = 0.1)
y_hat_HTavg['Holt_linear'] = fit1.forecast(len(test))
```

```python
plt.figure(figsize=(16,8))
plt.plot(train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_HTavg['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Count, y_hat_HTavg.Holt_linear))
print(rms)
```

### Winter's Holt Model

```python
y_hat_HWavg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['Count']) ,seasonal_periods=7 ,trend='add', seasonal='add',).fit()
y_hat_HWavg['Holt_Winter'] = fit1.forecast(len(test))
```

```python
plt.figure(figsize=(16,8))
plt.plot( train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_HWavg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Count, y_hat_HWavg.Holt_Winter))
print(rms)
```

### ARIMA

```python
y_hat_Aavg = test.copy()
fit1 = sm.tsa.statespace.SARIMAX(train.Count, order=(2, 1, 4),seasonal_order=(0,1,1,7)).fit()
y_hat_Aavg['SARIMA'] = fit1.predict(start="2013-11-1", end="2013-12-31", dynamic=True)
```

```python
plt.figure(figsize=(16,8))
plt.plot( train['Count'], label='Train')
plt.plot(test['Count'], label='Test')
plt.plot(y_hat_Aavg['SARIMA'], label='SARIMA')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.Count, y_hat_Aavg.SARIMA))
print(rms)
```

```python
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    acf1 = 0#acf(fc-test)[1]                      # ACF1
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse, 'acf1':acf1, 
            'corr':corr, 'minmax':minmax})
```
